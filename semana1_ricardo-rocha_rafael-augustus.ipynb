```python
import pandas as pd
import matplotlib.pyplot as plt
import math
volumes = [math.pi*0.01*(0.011**2), math.pi*0.02*(0.011**2), math.pi*0.03*(0.011**2), math.pi*0.04*(0.011**2), math.pi*0.05*(0.011**2), math.pi*0.06*(0.011**2), math.pi*0.07*(0.011**2), math.pi*0.08*(0.011**2)]

data = {
    "Nº": [1,2,3,4,5,6,7,8],
    "p(N)": [0.76, 0.71, 0.66, 0.61, 0.56, 0.51, 0.46, 0.41],
    "Altura Submersa (m)": [1/100, 2/100, 3/100, 4/100, 5/100, 6/100, 7/100, 8/100],
    "Volume Submerso (m³)": volumes
}

df = pd.DataFrame(data)

fig, ax = plt.subplots(figsize=(12, 3))
ax.axis('off')
table = ax.table(cellText=df.values, colLabels=df.columns, loc='center', cellLoc='center')
table.auto_set_font_size(False)
table.set_fontsize(10)
table.scale(1, 2)

plt.tight_layout()
plt.show()
```

```python
pesos = [0.76, 0.71, 0.66, 0.61, 0.56, 0.51, 0.46, 0.41]
volumes = [math.pi*0.01*(0.011**2), math.pi*0.02*(0.011**2), math.pi*0.03*(0.011**2), math.pi*0.04*(0.011**2), math.pi*0.05*(0.011**2), math.pi*0.06*(0.011**2), math.pi*0.07*(0.011**2), math.pi*0.08*(0.011**2)]
empuxos = []
densidades = []
soma_densidade = 0
for peso in pesos:
    empuxos.append(round(0.8 - peso, 2))
for empuxo in empuxos:
    print(f"Empuxo: {empuxo}")
for volume, empuxo in zip(volumes,empuxos):
    densidades.append(round(empuxo/(9.8*volume), 2))
for densidade in densidades:
    print(f"Densidade: {densidade}")
    soma_densidade += densidade
media_densidade = round(soma_densidade/(len(densidades)),2 )
print(f"Média da densidade: {media_densidade}")
```
